```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax);

plt.xlabel('Colums');
plt.ylabel('Missing Values');

plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books['description'].isna(), 1, 0).astype(int)
books['age_of_book']= 2025 - books['published_year']
```

```python
books# Check if 'age_of_the_book' exists
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt="0.2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})

heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
book_missing = books[~(books["description"]).isna() & ~(books["num_pages"]).isna() & ~(books["average_rating"]).isna() & ~(books["published_year"]).isna()]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing
```

```python
book_missing.loc[:, "description"] = book_missing["description"].fillna("").astype(str)
book_missing.loc[:, "words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing.loc[book_missing["words_in_description"].between(1,4), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(15,24), "description"]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"]>=25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words.loc[:, "title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],  # Use title if subtitle is NaN
    book_missing_25_words["title"] + ": " + book_missing_25_words["subtitle"]  # Concatenate title & subtitle
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words.loc[:, "tagged_description"] = book_missing_25_words[['isbn13','description']].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(book_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1).to_csv("books_cleaned.csv", index=False))
```
